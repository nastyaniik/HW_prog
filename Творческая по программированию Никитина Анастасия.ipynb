 Vega-altair - это библиотека визуализации на Python, основанная на javascript-библиотеках vega и vega-lite
 Vega и vega lite - это декларативные языки программирования, в которых вы задаете свойства графика в формате JSON, и он строит графики на основе этого, используя Canvas или SVG.
 
 
 
 JSON — текстовый формат обмена данными, основанный на JavaScript
 Canvas — элемент HTML5, предназначенный для создания растрового двухмерного изображения при помощи скриптов на языке JavaScript
 HTML5 - язык для структурирования и представления содержимого
 SVG (Scalable Vector Graphics — «масштабируемая векторная графика») — это вид графики, который создают с помощью математического описания геометрических примитивов (линий, кругов, эллипсов, прямоугольников, кривых и так далее), которые и образуют все детали будущего изображения.
 
 
 
 Vega-altair предлагает мощную и краткую грамматику, которая позволяет быстро создавать широкий спектр статистических визуализаций. 

```python
Вот пример использования API для визуализации набора данных с помощью интерактивной точечной диаграммы:
    
import altair as alt 
#загрузите образец набора данных в виде Pandas DataFrame
from vega_datasets import data
cars = data.cars()

# создание диаграммы
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).interactive()


Pandas — это программная библиотека на языке Python для обработки и анализа данных.
DataFrame — основной тип данных в Pandas, вокруг которого строится вся работа.

Вы объявляете связи между столбцами данных и каналами визуального кодирования, такими как оси x, y и цвет. Остальные детали графика обрабатываются автоматически.
```

Vega-altair может провизуализировать практически все: от простых диаграмм до тематических исследований и интерактивных диаграмм.
Но вот еще один пример простой диаграммы, нарисованной с помощью Vega-altair:

```python
import altair as alt
import pandas as pd


#придаем каждому столбцу значение
source = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': [28, 55, 43, 91, 81, 53, 19, 87, 52]
})

# импортируем a и b как оси графика x и y, чтобы программа поняла как их правильно расположить
alt.Chart(source).mark_bar().encode(
    x='a',
    y='b'
)
```
